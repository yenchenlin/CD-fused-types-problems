```python
import numpy as np
from sklearn.linear_model.coordinate_descent import ElasticNet
```

```python
# Init data
np.random.seed(5)
X = np.random.rand(400000, 40)
y = np.random.rand(400000)
T = np.random.rand(5, 40)
```

```python
# Fit our ElasticNet
clf = ElasticNet(alpha=1e-8, l1_ratio=1.0, precompute=False)
clf.fit(X, y)

# Predict targets
pred = clf.predict(T)
print "--------------"
print "pred: ", pred
```

```python
# Turn data into float32
X = np.float32(X)
y = np.float32(y)
T = np.float32(T)
```

```python
# Fit our ElasticNet
clf = ElasticNet(alpha=1e-8, l1_ratio=1.0, precompute=False)
clf.fit(X, y)

# Predict targets
pred = clf.predict(T)
print "--------------"
print "pred: ", pred
```
